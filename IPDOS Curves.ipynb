```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Data sets for FeV3 and Fe3V alloys

d1 = pd.read_csv('TEST11.csv')
d2 = pd.read_csv('Fe50V50_measured_DOS.csv')

fig1, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,3))

#FIRST GRAPH
axes[0].plot(d1)
axes[0].set_title('TEST1')
axes[0].set_xlabel('Energy')
axes[0].set_xlim([0.0,1.0])
axes[0].set_ylabel('INTERACTION IPDOS')
axes[0].set_ylim([-1.0,0.07])

#THIRD GRAPH 
axes[1].plot(d2)
axes[1].set_title('Fe50 V50')
axes[1].set_xlabel('Energy')
axes[1].set_xlim([0.0,1.0]) #Tried rescaling energy from 0-1... fail
axes[1].set_ylabel('Phonon DOS')
axes[1].set_ylim([-1.0,0.07])
```
